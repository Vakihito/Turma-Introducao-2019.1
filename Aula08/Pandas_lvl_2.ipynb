# 4 - Groupby (split-apply-combine)

http://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html

## Q: Qual é a média de gols por copa?

```python
# por copa 1
matches.groupby('Year')[['HomeTeamGoals', 'AwayTeamGoals']].sum().sum(axis = 1) / matches.groupby('Year').size()
```

```python
# por copa 2
matches.groupby('Year')[['HomeTeamGoals', 'AwayTeamGoals']].mean().sum(axis = 1)
```

## Q: Qual é a média de gols por país sede?

```python
# Repete país sede?
cups.Country.value_counts()
```

```python
mcols = ['Year', 'HomeTeamGoals', 'AwayTeamGoals']
ccols = ['Year', 'Country']
pd.merge(matches[mcols], cups[ccols], on = 'Year')
```

```python
mcols = ['Year', 'HomeTeamGoals', 'AwayTeamGoals']
ccols = ['Year', 'Country']
pd.merge(matches[mcols], cups[ccols], on = 'Year').groupby('Country')[['HomeTeamGoals', 'AwayTeamGoals']].mean().sum(axis = 1)
```

## Q: Qual é o país com maior público médio nas últimas 3 copas

```python
last3 = cups['Year'].sort_values().values[-3:]
```

```python
cols = ['Year', 'Attendance']
matA = matches[cols + ['HomeTeamName']].rename(columns = {'HomeTeamName' : 'TeamName'})
matB = matches[cols + ['AwayTeamName']].rename(columns = {'AwayTeamName' : 'TeamName'})

mats = pd.concat([matA, matB], axis = 0)
```

```python
mat3 = mats.query('Year in @last3')
```

```python
mat3.groupby('TeamName')['Attendance'].mean().sort_values()
```

```python
mat3.groupby('TeamName')['Attendance'].mean().idxmax()
```

# 5 - Datetime e Timedelta

https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html

- Date times: A specific date and time with timezone support. Similar to datetime.datetime from the standard library.
        - dtype : datetime64[ns] or datetime64[ns, tz]
- Time deltas: An absolute time duration. Similar to datetime.timedelta from the standard library.
        - dtype : timedelta64[ns]
        
### Datetime - Datetime = Timedelta
### Datetime + Timedelta = Datetime

## Q: Quantos dias desde a última vez que um time jogou?

```python
matches.Datetime.head()
```

```python
matches.Datetime.str.split().str[1].unique()
```

```python
(matches.Datetime.str.split().str[1].isin({'June', 'July'})).sum()
```

```python
matches['Datetime'] = matches.Datetime.str.replace('June', 'Jun').str.replace('July', 'Jul')
```

```python
matches['Datetime'] = pd.to_datetime(matches.Datetime, format = '%d %b %Y - %H:%M ')
```

```python
matches['Datetime'].head()
```

```python
cols = ['Datetime']
matA = matches[cols + ['HomeTeamName']].rename(columns = {'HomeTeamName' : 'TeamName'})
matB = matches[cols + ['AwayTeamName']].rename(columns = {'AwayTeamName' : 'TeamName'})
```

```python
mats = pd.concat([matA, matB], axis = 0)
```

```python
mats.head()
```

```python
mats.sort_values('Datetime', inplace = True)
```

```python
ult = mats.groupby('TeamName')['Datetime'].nth(-1)
ult.head()
```

```python
(pd.datetime.today() - ult) // pd.Timedelta(1, 'D') # or np.timedelta64(1, 'D')
```
